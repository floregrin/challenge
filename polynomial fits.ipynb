```python
def plotinterp1DExtrapolate(x,y,extraDates,outdir):    
    plt.plot(x,y,label='cases')
    plt.title(state+' NeW CASES ROLLED AVERAGE 7D')
    
    plt.plot(x,abs(scipy.interpolate.interp1d(x[:-extraDates],y[:-extraDates],fill_value='extrapolate')(x)),label='interp1D')
    plt.ylabel('value')
    plt.xlabel('days')
    plt.legend()
    plt.savefig(outdir+state+'casesRolled7Dinterp1Dextrapolate'+str(extraDates)+'.png')
    plt.close()
    return
```

```python
def plotpolyExtrapolate(x,y,extraDates,outdir,deg):    
    plt.plot(x,y,label='cases')
    plt.title(state+' NeW CASES ROLLED AVERAGE 7D')
    
    yhat=np.poly1d(np.polyfit(x[:-extraDates], y[:-extraDates], deg))
    plt.plot(x,abs(yhat(x)),label='poly'+str(deg))
    plt.ylabel('value')
    plt.xlabel('days')
    plt.legend()
    plt.savefig(outdir+state+'casesRolled7DpolyExtrap'+str(deg)+'_'+str(extraDates)+'.png')
    plt.close()
    return
```

```python
def plotsplineExtrapolate(x,y,extraDates,outdir):    
    plt.plot(x,y,label='cases')
    plt.title(state+' NeW CASES ROLLED AVERAGE 7D')
   
    spl = splrep(x[:-extraDates], y[:-extraDates])
    y2 = splev(x, spl)
    
    plt.plot(x,abs(y2),label='spline1d')
    plt.ylabel('value')
    plt.xlabel('days')
    plt.legend()
    plt.savefig(outdir+state+'casesRolled7DsplineExtrap'+'_'+str(extraDates)+'.png')
    plt.close()
    return
```
