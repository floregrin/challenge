```python
dfs = []
for fn in fns:
    dfs.append(pd.read_csv(fn))
```

```python
dfs[0].head()
```

```python
for df in dfs:
    df['date'] = pd.to_datetime(df['date'])
```

```python
for i,df in enumerate(dfs):
    df = df[['country','date', 'cli_se', 'cli_se_unw', 'percent_cli', 'percent_cli_unw']]
    dfs[i] = df
```

```python
dfr = pd.concat(dfs)
```

```python
import numpy as np
graph_df = np.empty(len(dfs), dtype=object)
```

```python
rollingDays=7
column = 'percent_cli_unw'
#for i, df in enumerate(dfs):
graph_df = pd.DataFrame()
df = dfr
for region in df['country'].unique()[2:]:
    region = region.replace(" ", "")
    print(region)
    region_df = df.copy()[df['country']==region]
    region_df.set_index('date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f"{region}_val" + str(rollingDays)+"ma"] = region_df[column].rolling(rollingDays).mean()

    if graph_df.empty:
        graph_df = region_df[[f"{region}_val"+str(rollingDays) + "ma"]]  # note the double square brackets!
    else:
        graph_df = graph_df.join(region_df[f"{region}_val"+ str(rollingDays) + "ma"])
```

```python
try:
    os.mkdir('results/')
except:
    pass
```

```python
graph_df.dropna(axis=1).to_csv('results/'+column+'_ notna_'+ str(rollingDays) + 'D.csv')
```

```python
import matplotlib.pyplot as plt

graph_df.plot(figsize=(8,5), legend=False, title=column+ ' countrys 7day average')
plt.savefig('results/'+column+'7Dnotna.png')
```
